# Dados √önicos

```python
df2 = pd.DataFrame({
    'A': 1,
    'B': pd.Timestamp('20130102'),
    'C': pd.Series(1, index=list(range(4)), dtype='float32'),
    'D': np.array([3] * 4, dtype='int32'),
    'E': pd.Categorical(['test', 'train', 'test', 'train']),
    'F': 'Python',
    'G': [2, 2, 4, 4],
    'H': [np.nan, 2, 4, np.nan]
})
df2
```

```python
df2.nunique() # Retorna s√©rie com n√∫mero de valores distintos. Pode ignorar valores NaN
```

```python
df2.nunique(dropna=False) # NaN agora √© contabilizado
```

```python
df2.nunique(axis=1) # Retorna s√©rie com n√∫mero de valores distintos por linha
```

```python
df2.nunique(axis=1, dropna=False)
```

# Removendo Duplicatas

```python
df2
```

```python
df2.drop_duplicates() # Remove linhas duplicadas. Avalia se existe alguma linha igual a outra.
```

```python
df2.drop_duplicates(subset=['G']) # Avalia a coluna G e remove os duplicados
```

```python
df2.drop_duplicates(subset=['G'], keep='last') # Mant√©m a √∫ltima ocorr√™ncia
```

```python
df2.drop_duplicates(subset=['G'], keep=False, )
```

# Ordena√ß√£o dos Dados

```python
df = pd.DataFrame({
    'Col1': ['A','A','B', np.nan, 'D', 'C'],
    'Col2': [2,1,9,8,7,4],
    'Col3': [0,1,9,4,2,3],
})
df
```

```python
df.sort_values(by='Col1', ascending=True)
```

```python
df.sort_values(by='Col1', ascending=False)
```

```python
df.sort_values(by='Col2')
```

```python
df.sort_values(by=['Col3','Col1']) # Ordena primeiro pela coluna Col3 e depois pela Col1
```
