# Boxplot

```python
np.random.seed(seed=0)

x = np.random.randn(1000)
y = np.random.randn(100)
z = np.random.randn(10)
```

```python
fig, ax = plt.subplots()
ax.boxplot(
    (x,y,z), # conjunto de dados
    vert=False, # plotagem horizontal
    showmeans=True, # mostra a média
    meanline=True, # linha da média
    labels=('X', 'Y', 'Z'), # labels
    patch_artist=True, # preencher com cores as caixas
    medianprops={'linewidth': 2, 'color': 'purple'}, # cor da linha da mediana
    meanprops={'linewidth': 2, 'color': 'red'} # cor da linha da média
)
plt.show()
```

# Salvar figuras

```python
fig, ax = plt.subplots()
ax.boxplot(
    (x,y,z), # conjunto de dados
    vert=True, # plotagem horizontal
    showmeans=True, # mostra a média
    meanline=True, # linha da média
    labels=('X', 'Y', 'Z'), # labels
    patch_artist=True, # preencher com cores as caixas
    medianprops={'linewidth': 2, 'color': 'purple'}, # cor da linha da mediana
    meanprops={'linewidth': 2, 'color': 'red'} # cor da linha da média
)
plt.savefig('images/boxplot.png', transparent = False); # Deve ficar antes do show(). Salva a imagem do ultimo gráfico plotado.
plt.show()
```
